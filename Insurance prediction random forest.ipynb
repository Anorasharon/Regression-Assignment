```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_female', 'sex_male','smoker_no', 'smoker_yes' ]]
```

```python
dependent=dataset[['charges']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.31,random_state=0)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
#regressor=RandomForestRegressor(n_estimators=50,random_state=0,criterion='squared_error')
#regressor=RandomForestRegressor(n_estimators=50,random_state=0,criterion='absolute_error')
#regressor=RandomForestRegressor(n_estimators=50,random_state=0,criterion='friedman_mse')
regressor=RandomForestRegressor(n_estimators=50,random_state=0,criterion='poisson')
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```
